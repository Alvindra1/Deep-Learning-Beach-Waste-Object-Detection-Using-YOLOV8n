```python
from ultralytics import YOLO
import pandas as pd
import os
```

```python
model = YOLO('./best.pt')
print("Load Fine Tuned Model")
```

```python
results = model.val(data="plastic_coco.yaml", verbose=True, save=True, project="evaluate_test", name="results")
```

```python
metrics = results.summary()
metrics_df = pd.DataFrame([metrics])
metrics_csv_path = os.path.join("evaluate_test", "yolo_metrics.csv")
metrics_df.to_csv(metrics_csv_path, index=False)
```
